<h2>Introduction</h2>

In the city of Curitiba, there are some venues that offers a service called "café colonial", a kind of "all you can eat" coffee/bakery service. And the focus of this work is to compare then in order to decide which one is the best to a possible client.

Despite being a tasty service, it can be very hard to classify then by actually experimenting them all (very costly and fattening!), and as it is a food service, personal taste and subjective factors are also very relevant. So, to solve this matter, there will be created groups with the venues accordingly to some similar characteristics and the recommend the group as a whole, instead of only one sugestion.

<h2>Data</h2>
The data for this work is obtained through the foursquare functionalities: venues from Curitiba and nearby region that have "café colonial", their locations, evaluations and if there are plants stores nearby.

<h2>Methodology</h2>

The venues will be evaluated and grouped according to the following criteria:
<ol>
<li>If it opens on Sundays and close at a late time (until 18-19h)</li> 
    The best day to go to an "all you can eat" is Sunday, and once you get in, you have to stay there until a very late time, eating. So the data will receive values 0 (don't open) or 1 (open)
<li>The quality</li>    
    The rating from foursquare
<li>The quality</li>    
    The price from foursquare
<li>Amount of a plant/garden venues nearby</li>
    The count of plants stores that within a 1km radius from the venue. Going to a plant/garden store, enjoy myself while at it and then go to a <b>café colonial</b> is the perfect plan to a perfect day.
    Ok, this criteria is very subjective since it is based on my opinion, but nevertheless, is interesting and uses location based data to show that i am able do work with it
    <li>Be on Curitiba's region</li>
   Can be Curitiba or it's neighbouring regions
</ol>

Once all this data is gathered, the group classification will be done using python's library KMeans from sklearn.cluster, dividing the data in 5 clusters that will be further classified

<h2>Discussion</h2>
From the results, we can analyse the data and in order to define why the cluster were divided like this by this clustering method.

And the easiest way to do that is graphically: 

```python
graficoCafe = df.groupby('cluster').aggregate(np.mean).reset_index()[['plants','price tier','rating','late sunday']]
x = np.arange(5)
ys = [i+x+(i*x)**2 for i in range(5)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
markers_colors = []
grupsCC = {0:"Cluster 0",1:"Cluster 1",2:"Cluster 2",3:'Cluster 3',4:'Cluster 4'}
plt.bar([1,2,3,4,5],graficoCafe['plants'], width = 0.9, color = rainbow, label='Plants')
plt.xticks([1,2,3,4,5], grupsCC.values(), rotation=90)
plt.title('Plants stores nearby')
plt.show()
```

Since cluster 2 in the one with, in average, more plants venues nearby. But this plot can be misleading, since there is one element in cluster 2 without any cases of plants venues nearby 

```python
plt.bar([1,2,3,4,5],graficoCafe['late sunday'], width = 0.9, color = rainbow, label='Plants')
plt.xticks([1,2,3,4,5], grupsCC.values(), rotation=90)
plt.title('Opens until late at sunday')
plt.show()
```

This plot shows that the clusters 0 and 4 do not open on sundays at late hours, while all the others do. It is a very clear separation in this aspect

```python
plt.bar([1,2,3,4,5],graficoCafe['price tier'], width = 0.9, color = rainbow, label='Plants')
plt.xticks([1,2,3,4,5], ['Cluster 0', 'Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4'], rotation=90)
plt.title('Average price tier')
plt.show()
```

We can see that Cluster 2 is the most expansive in average, while cluster 1 and 4 arethe cheaper ones

```python
plt.bar([1,2,3,4,5],graficoCafe['rating'], width = 0.9, color = rainbow, label='Plants')
plt.xticks([1,2,3,4,5], grupsCC.values(), rotation=90)
plt.title('Average rating')
plt.show()
```

All the ratings are pretty close, but luster 3 takes the lead here as the one with better ratings

```python
plt.bar([1,2,3,4,5], (graficoCafe['rating']/graficoCafe['price tier']), width = 0.9, color = rainbow, label='Plants')
plt.xticks([1,2,3,4,5], grupsCC.values(), rotation=90)
plt.title('Ratings/price tier ratio')
plt.show()
```

This is a simple divisions between ratings and price ratio, where we can obtain a simple 'cost-benefit'ratio
Cluster 4 is the one in the lead here.

With this information, we can define each cluster accordingly to it's mains features and rename then, making easy to an possible costumer decide which group of venues would be more interesting to him/her visit:

So cluster 0 will became the "Well rounded group", due to it's overall average positions<br>
cluster 1: "Cheapest group", the group with the cheapest price (and somewhat low ratings)<br>
cluster 2: "Fanciest group", the group with the most expansive services.<br>
